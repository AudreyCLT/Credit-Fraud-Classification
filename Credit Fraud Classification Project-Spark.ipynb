```python
df.head()
```

```python
required_features = df.drop('Class').columns
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
assembler = VectorAssembler(inputCols=required_features, outputCol='features')
transformed_data = assembler.transform(df)
```

```python
(training_data, test_data) = transformed_data.randomSplit([0.8,0.2])

training_data.cache()
```

```python
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator

def run_randomForests(training_data, test_data, features):
    rf = RandomForestClassifier(labelCol='Class', 
                                featuresCol=features,
                                maxDepth=5)
    print('Train set')
    model = rf.fit(training_data)
    predictions = model.transform(training_data)
    evaluator = BinaryClassificationEvaluator(
        labelCol="Class", rawPredictionCol="prediction", metricName="areaUnderROC")
    recallByLabel = evaluator.evaluate(predictions)
    print("Train Area Under ROC = %g" % (recallByLabel))
    print('Test set')
    model = rf.fit(test_data)
    predictions = model.transform(test_data)
    evaluator = BinaryClassificationEvaluator(
        labelCol="Class", rawPredictionCol="prediction", metricName="areaUnderROC")
    recallByLabel = evaluator.evaluate(predictions)
    print("Test Area Under ROC = %g" % (recallByLabel))
    
    
```

```python
run_randomForests(training_data, test_data)
```

However, this 0.85 score is not indicative of the performance of the random forest classifier.  The reason for that is because remember that majority of the data is classified as class 0, and only a small portion is classified as class 1.  That means that, if the model predict the class to be 0 for all data, we would still get a high score.

### Balance Imbalance Set

As the problem we have seen so far is that the model has problem with identifying fraudulent transactions due to the overwhelming portion of non-fraudulent transactios.  To help with this, we will resample the data set to ensure a balance training set can be sampled.

```python
from pyspark.sql.functions import col, explode, array, lit


major_df = df.filter(col("class") == 0)
minor_df = df.filter(col("class") == 1)
ratio=major_df.count()/minor_df.count()

print("Number of non-fraudulent data: {}".format(major_df.count()))
print("Number of fraudulent data: {}".format(minor_df.count()))
```

```python
# under sampling
sampled_majority_df = major_df.sample(False, 1/ratio)
undersampled_df = sampled_majority_df.unionAll(minor_df)
undersampled_df.show()


```

```python
# over sampling

a = range(int(ratio))
# duplicate the minority rows
oversampled_transformed_df = minor_df.withColumn("dummy", explode(array([lit(x) for x in a]))).drop('dummy')
# combine both oversampled minority rows and previous majority rows 
oversampled_df = major_df.unionAll(oversampled_transformed_df)
oversampled_df.show()
```

```python
assembler = VectorAssembler(inputCols=required_features, outputCol='features')
undersampled_df = assembler.transform(undersampled_df)

(undersampled_training, undersampled_testing) = undersampled_df.randomSplit([0.8,0.2])
run_randomForests(undersampled_training, undersampled_testing)
```

## Putting the above together 

```python
# load dataframe
df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load("creditcard.csv")\
    .coalesce(5)\

# drop duplicates
df = df.dropDuplicates()

# undersample data
major_df = df.filter(col("Class") == 0)
minor_df = df.filter(col("Class") == 1)
ratio=major_df.count()/minor_df.count()
sampled_majority_df = major_df.sample(False, 1/ratio)
undersampled_df = sampled_majority_df.unionAll(minor_df)

required_features = df.drop('Class').columns

assembler = VectorAssembler(inputCols=required_features, outputCol='features')
undersampled_df = assembler.transform(undersampled_df)

# bucketing
for col in df.columns:
    print(col)
    if col=='Class':
        break
    col_min=float(describe[col][3])
    col_max=float(describe[col][4])
    array=[float(x) for x in np.linspace(start = col_min, stop = col_max, num = 10)]
    print('test')
    bucketer = Bucketizer().setSplits(array).setInputCol(col)
    df=bucketer.transform(df)
        
# scaling
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures",
                        withStd=True, withMean=False)
scalerModel = scaler.fit(undersampled_df)
undersampled_df = scalerModel.transform(undersampled_df)

# splitting data into training and testing set
(undersampled_training, undersampled_testing) = undersampled_df.randomSplit([0.8,0.2])
```

```python
run_randomForests(undersampled_training, undersampled_testing, 'scaledFeatures')
```
